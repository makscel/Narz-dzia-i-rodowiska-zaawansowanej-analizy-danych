```python
import pandas as pd
from sklearn.metrics import adjusted_rand_score
import seaborn as sns
import matplotlib.pyplot as plt

# Wczytanie danych
data_with_groups = pd.read_csv('my-penguins-with-groups.csv', sep=',', encoding='utf-8')

# Macierz porównawcza
comparison_matrix = pd.crosstab(data_with_groups['species'], data_with_groups['group'])

# Wyświetlenie macierzy
plt.figure(figsize=(10, 6))
sns.heatmap(comparison_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Comparison Matrix of Species vs Group')
plt.xlabel('Group')
plt.ylabel('Species')
plt.show()

# Ocena jakości grupowania
ari_score = adjusted_rand_score(data_with_groups['species'], data_with_groups['group'])
print(f'Adjusted Rand Index: {ari_score}')
```
